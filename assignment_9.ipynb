```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python


data = pd.DataFrame({
    'RAM':[4,6,8,12,16],
    'Storage':[64,128,256,256,512],
    'Battery':[4000,4500,5000,5000,6000],
    'Price':[10000,15000,22000,30000,45000]
})

X = data[['RAM','Storage','Battery']]
y = data['Price']

print(X)
print(y)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1)
lasso.fit(X,y)

print("Lasso Coefficients")
print(lasso.coef_)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)
ridge.fit(X,y)

print("Ridge Coefficients")
print(ridge.coef_)
```

Lasso may set some coefficients to zero, effectively removing features, while Ridge keeps all coefficients but reduces their values.


```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(alpha=1,l1_ratio=0.5)
elastic.fit(X,y)

print("ElasticNet Coefficients")
print(elastic.coef_)
```

ElasticNet combines both L1 and L2 regularization, so its coefficients are usually between those of Lasso and Ridge.


```python
print("Features Kept:")
for i in range(len(X.columns)):
    if lasso.coef_[i] != 0:
        print(X.columns[i])

print("\nFeatures Dropped:")
for i in range(len(X.columns)):
    if lasso.coef_[i] == 0:
        print(X.columns[i])
```

Lasso regression kept all three features (RAM, Storage, and Battery) because their coefficients were non-zero. Therefore, no features were removed by the feature selection process.
