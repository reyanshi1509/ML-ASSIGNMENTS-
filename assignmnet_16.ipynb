```python
tp = 50
tn = 120
fp = 10
fn = 20

accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
def print_confusion_matrix(tp, tn, fp, fn):
    print("TP =", tp)
    print("TN =", tn)
    print("FP =", fp)
    print("FN =", fn)

print_confusion_matrix(50, 120, 10, 20)
```
