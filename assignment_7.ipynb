```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
actual_delivery = np.array([25, 30, 28, 35, 40, 32, 27, 38, 29, 34])

predicted_delivery = np.array([24, 32, 27, 36, 39, 31, 29, 37, 30, 33])

mse = mean_squared_error(actual_delivery, predicted_delivery)
rmse = np.sqrt(mse)

print( mse)
print( rmse)
```

```python


actual_prices = np.array([15000, 18000, 22000, 25000, 30000])
predicted_prices = np.array([14800, 17500, 22500, 24800, 29500])

mae = mean_absolute_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)

print("MAE:", mae)
print("R² Score:", r2)
```

```python
import matplotlib.pyplot as plt

actual = [4,5,3,4,5,2,3,4,5,4,3,5,2,4,5,3,4,5,3,4]
predicted = [4,4,3,5,4,2,4,4,5,3,3,5,3,4,4,3,5,4,3,4]

residuals = []

for i in range(len(actual)):
    residuals.append(actual[i] - predicted[i])

plt.plot(residuals, 'o')
plt.xlabel("Movie Number")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.show()

```

the residuals are scattered around zero, indicating that the model errors are relatively random.

Model B is preferred because it has a lower RMSE (6 minutes) than Model A (8 minutes), which means its predictions are closer to the actual delivery times. Although Model A has a higher R² score, lower prediction error is more important for delivery time prediction, making Model B the better choice.


```python


actual = [60,72,80,65,90,55,78,85]
model_a = [61,70,81,64,89,57,77,84]
model_b = [55,82,68,75,98,45,88,95]

for name, pred in [("Model A", model_a), ("Model B", model_b)]:
    print("\n", name)
    print("MSE:", mean_squared_error(actual, pred))
    print("RMSE:", np.sqrt(mean_squared_error(actual, pred)))
    print("MAE:", mean_absolute_error(actual, pred))
    print("R²:", r2_score(actual, pred))
```

The model with lower MSE, RMSE, and MAE and higher R² performs better. Overfitting occurs when a model performs very well on training data but poorly on test data.
